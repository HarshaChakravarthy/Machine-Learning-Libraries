### Merge, Join & Concate

#### Merge

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"H:\Harsha\Self Study\Python\Pandas Practice\LOTR.csv")
df2 = pd.read_csv(r"H:\Harsha\Self Study\Python\Pandas Practice\LOTR 2.csv")
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'cross')
```

#### Join

```python
df1.join(df2, how = 'outer', on = 'FellowshipID' ,lsuffix = 'Left', rsuffix = 'Right')
```

```python
df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how = 'outer', lsuffix = 'Left', rsuffix = 'Right')
```

#### Concatenate

```python
pd.concat([df1, df2], join = 'outer')
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
```

### Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_excel(r"H:\Harsha\Self Study\Python\Pandas Practice\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('abc./_')
df
```

```python
df[['Strees_Address', 'City', 'Zipcode']] = df['Address'].str.split(',', n=2, expand = True)
df
```

```python
df = df.drop(columns = 'Address')
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^A-Za-z0-9]', "", regex = True)
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--', "")
```

```python
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('Na--', "")
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('No', 'N')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
        
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == "":
        df.drop(x, inplace = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Y', 'Yes')
df['Paying Customer'] = df['Paying Customer'].replace('N', 'No')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N', 'No')
df
```

```python
df
```

```python
df = df.reset_index(drop = True)
```
