```python
import matplotlib.pyplot as plt
```

## Line Chart

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y, c='b', linestyle = ':', linewidth = '2', marker = '*', markersize = 8)
plt.title('Simple Line Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 20, 25, 30]

plt.plot(x,y1, c = 'r', label='Red', linestyle = '-')
plt.plot(x,y2, c = 'b', label = 'Blue', linestyle = '--')

plt.legend()
plt.show()
```

## Scatter Plot

```python
x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 35, 45]

plt.scatter(x,y, color = 'g', marker = '*', s = 100, alpha = 0.5)
plt.title('Basic Scatter Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## Bar Chart

```python
x = [10,20,15,25,30]
y = ['A','B','C','D','E']

plt.bar(x,y)
plt.title('Simple Bar Chart')
plt.xlabel('Variables')
plt.ylabel('Categories')
plt.show()
```

```python
plt.barh(x,y, color = 'g')
```

```python
plt.bar(x,y, color=['red','green', 'blue','orange'], edgecolor = 'black', linewidth = 2)
plt.show()
```

```python
# Histogram

import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins = 30, edgecolor = 'black', color = 'green')
plt.title('Basic Histogram Chart')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

## Pie Chart

```python
# Basic Pie Chart

Fruits = ['Apple', 'Banana', 'Grapes', 'Mango']
size = [15,25,20,40]

plt.pie(size, labels = Fruits)
plt.title('Demand of Fruits')
plt.show()
```

```python
# Adding Features Pie Chart

Fruits = ['Apple', 'Banana', 'Grapes', 'Mango']
size = [15,25,20,40]

colors = ['red','blue','orange','green']
explode = [0.2,0,0,0]
plt.pie(size, labels = Fruits, colors = colors, explode = explode, autopct = '%1.1f%%', startangle = 10)


plt.title('Demand of Fruits')
plt.show()
```

## Sub Plots

```python
plt.figsize = (10,5)

plt.subplot(1,2,1)
plt.plot([1,2,3],[10,20,30])

plt.show()
```

```python
plt.figsize = (8,8)

plt.subplot(1,2,1)
plt.bar(['A','B','C'],[10,20,30])

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Line Plot")

plt.subplot(1, 2, 2)
plt.bar(["A", "B", "C"], [10, 20, 30])
plt.title("Bar Chart")
plt.tight_layout()

plt.show()
```

## Adding Legends

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y,color = 'g', label = 'Green Line')
plt.title('Line Chart adding Legends')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc = 'upper left')
# plt.legend(loc = 'upper right')
# plt.legend(loc = 'lower left')
# plt.legend(loc = 'lower right')
plt.legend(loc = 'best')

plt.show()
```

## Adding Annotations

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]

plt.plot(x, y1, label="Line 1", color="blue")

plt.annotate("Peak Point", xy=(5, 50), xytext=(3, 45),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.legend()
plt.show()
```
