## Basic Plots (Line, Scatter, Bar)

### Line Chart

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='mpg', y = 'hp', data=data)
plt.title('Miles per Gallon vs Horse Power')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Horse Power')
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x = 'mpg', y = 'wt', data=data, hue = 'cyl', palette = 'Set1')
plt.title('Miles per Gallon vs Car Weight')
plt.xlabel('Miler per Gallon')
plt.ylabel('Car Weight')
plt.show()
```

### Bar Plot

```python
plt.figure(figsize=(8,6))
sns.barplot(x = 'mpg', y = 'hp', data=data)
plt.title('Miles per Gallon Vs Horse Power')
plt.xlabel('Miles per Gallon')
plt.ylabel('Horse Power')
plt.show()
```

## Statistical Plots (Box plot, Violin Plot, Histogram, KDE Plot)

### Box Plot

```python
plt.figure(figsize=(4,6))
sns.boxplot(x='cyl',y='mpg',data=data)
plt.title('MPG Distribution Across Cylinder Types')
plt.xlabel('Cylinder')
plt.ylabel('Miles per Gallon')
plt.show()
```

### Violin Plot

```python
plt.figure(figsize=(4,4))
sns.violinplot(x='cyl',y='mpg',data=data, inner='quartile')
plt.title('MPG Distribution Across Cylinder Types')
plt.xlabel('Cylinder')
plt.ylabel('Miles per Gallon')
plt.show()
```

### Histogram

```python
plt.figure(figsize=(8,6))
sns.histplot(data['mpg'],bins = 25,color='green')
plt.title('Distribution of Miles per Gallon')
plt.xlabel('Miles per Gallon')
plt.ylabel('Count')
plt.show()
```

```python
# Adding KDE for the histogram

plt.figure(figsize=(8,6))
sns.histplot(data['mpg'],bins = 15,color='green', kde = True)
plt.title('Distribution of Miles per Gallon')
plt.xlabel('Miles per Gallon')
plt.ylabel('Count')
plt.show()
```

### KDE Plot

```python
sns.kdeplot(data['mpg'], color= 'red', fill = True)
plt.show()
```

```python
# KDE plot (Just for example)

sns.kdeplot(x = 'hp', y = 'mpg', data = data)
plt.show()
```

## Advanced Plots (Heatmaps, Pairplots, FacetGrid)

### Heatmap

```python
plt.figure(figsize=(10,8))
numeric = data.select_dtypes(include = ['number'])
sns.heatmap(numeric.corr(), annot = True, cmap = 'coolwarm')
plt.title('Heatmap')
plt.show()
```

### Pair Plot

```python
sns.pairplot(data=data)
```

```python
sns.pairplot(data, hue = 'cyl')
```

### FacetGrid

```python
grid = sns.FacetGrid(data, col = 'cyl')
grid.map_dataframe(sns.scatterplot, x = 'hp', y = 'mpg')
plt.show()
```

```python
# Using the same example for Barplot
grid = sns.FacetGrid(data, col='cyl')
grid.map_dataframe(sns.barplot, x = 'mpg', y = 'hp')
plt.show()
```
